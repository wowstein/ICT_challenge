```python
import docx
```

```python
def getText(filename):
    doc = docx.Document(filename)
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text)
    return '\n'.join(fullText)
```

```python
text = getText('/content/pride-and-prejudice.docx')
```

```python
len(text)
```

```python
import matplotlib.pyplot as plt
import wordcloud
import string
import numpy as np
```

```python
#Punctuation removal
punc_removed =''
for word in text:
  if word not in string.punctuation:
    punc_removed += word
```

```python
len(punc_removed)
```

```python
import nltk
nltk.download('punkt')
```

```python
word = nltk.word_tokenize(punc_removed)
```

```python
text_f = ""
for i in word:
  text_f = text_f + i + " "
```

```python
from wordcloud import WordCloud
```

```python
%pip install --upgrade Pillow
```

```python
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text_f)
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig("PriandPre_wordcloud.png", format="png")
plt.show()
```
